```python
# ensure annotation
def get_prodict(x:int,y:int) -> int:
    return x*y
```

```python
get_prodict(x=5,y=7)
```

```python
get_prodict(x=5,y='7')
```

```python
from ensure import ensure_annotations

# ensure annotation
@ensure_annotations
def get_prodict(x:int,y:int) -> int:
    return x*y
```

```python
get_prodict(x=5,y=7)
```

```python
get_prodict(x=5,y='7')
```
